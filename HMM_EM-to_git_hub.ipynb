```python
>>> import numpy as np
>>> import pandas as pd
>>> from hmmlearn import hmm
from sklearn.preprocessing import scale
import copy 
>>> np.random.seed(42)
```

```python
#loading a csv file with the data.
#each line in the table represenr a neighborhood per time period.

df = pd.read_csv('dfall2.csv').fillna(1)

####כאן אני אוסף את כל הפיצ'רים מסוג סטרינג והופך אותם לדאמיז כדי שאוכללהדביק אותם אחר כך בלי שיפריעו לי עם מילוי הערכים החסרים  

categorials = df.groupby(['msoa11cd']).agg({
    'msoa11nm':'first', 'rgn16nm':'first', 'metro' :'first',
    'RUC':'first','lad11cd':'first','metru_function':'first'})

categorials_dumm = pd.concat([pd.get_dummies(categorials[col]) for col in [
                                                                            'metru_function'
                                                                          ,'rgn16nm'
                                                                          ]], axis=1, keys=categorials.columns)

df.reset_index(inplace=True)
df=df.drop([ 'index',
           #    '%New','%Old', '%Lees', '%Free', '%Other', '%D', '%F', '%S', '%T',
               'metru_function', 'lad11cd', 'RUC', 'metro', 'rgn16nm',
               'Ref_area_price', 'Area Hectares',
                'All usual residents',
              'Density (number of persons per hectare)',
                'White',
               'Mixed/multiple ethnic groups', 'Asian/Asian British',
               'Black/African/Caribbean/Black British', 'Other ethnic group',
               'Average number of rooms per household',
                'Net annual income',
                'msoa11nm',
            #    'Mean_Price',
                'Transactions',
#                 'price_relative_to_ref_area'
               ],axis = 1)
##############################################333
df = pd.merge(df,categorials_dumm,left_on='msoa11cd', right_on='msoa11cd',how='left')
data=df.drop(['msoa11cd','Period'],axis = 1).to_numpy()

data
```

```python
neighborhoods = df.msoa11cd.unique()
print(neighborhoods.shape)

lengths=[]
for nei in neighborhoods:
    tempData=np.asarray(df[['price_relative_to_ref_area']].values[df['msoa11cd']==nei])
#     tempData=np.asarray(df[['price_relative_to_ref_area','%New','%Old','%Lees','%T','%F','%S','%D',
#                                    #       'Density (number of persons per hectare)', 'Net annual income',
                           
#                            ]].values[df['msoa11cd']==nei])
    #tempData=df['price_relative_to_ref_area'].values[df['msoa11cd']==nei].tolist()
   # data.append(tempData)
    lengths.append(len(tempData))
    print("_________________________________-",nei)
```

```python
#creating the HMM-EM model

model=hmm.GaussianHMM(n_components=7, n_iter=100, init_params="mcs").fit(data,lengths)

Y2=model.predict(data)

df['Y2']=Y2
df
```

```python
model=hmm.GaussianHMM(n_components=7, n_iter=100, covariance_type="diag",init_params="cm", params="cmt").fit(data,lengths)

Y3=model.predict(data)

df['Y3']=Y3
df
```
