```python
>>> import numpy as np
>>> import pandas as pd
>>> from hmmlearn import hmm
from sklearn.preprocessing import scale
import copy 
>>> np.random.seed(42)
```

```python
#loading a csv file with the data.
#each line in the table represenr a neighborhood per time period.

df = pd.read_csv('synthtec_sequance500.csv')

print(df)
####כאן אני אוסף את כל הפיצ'רים מסוג סטרינג והופך אותם לדאמיז כדי שאוכללהדביק אותם אחר כך בלי שיפריעו לי עם מילוי הערכים החסרים  
Y=df['CLASS']

df.reset_index(inplace=True)
df=df.drop([ 'index',
                'Unnamed: 0',
                'CLASS'
               ],axis = 1)
data=scale(df.drop(['msoa11cd','Period'],axis = 1))
##############################################333
df
```

```python
neighborhoods = df.msoa11cd.unique()
print(neighborhoods.shape)

lengths=[]
for nei in neighborhoods:
    tempData=np.asarray(df[['feature_1','feature_2','feature_3']].values[df['msoa11cd']==nei])

    lengths.append(len(tempData))
    print("_________________________________-",nei)
```

```python
model=hmm.GaussianHMM(n_components=7, n_iter=100, init_params="mcs").fit(data,lengths)
#model.fit(data)
Y2=model.predict(data)
#hmm.fit(data,lengths)
df['Y2']=Y2
df
```

```python
model=hmm.GaussianHMM(n_components=7, n_iter=100, covariance_type="diag",init_params="cm", params="cmt").fit(data,lengths)
#model.fit(data)
Y3=model.predict(data)
#hmm.fit(data,lengths)
df['Y3']=Y3
df
```

```python
df['Y_true']=Y
df
```

```python
df.to_csv('HMM_EM_syntisized.csv')
```

```python
from sklearn import metrics

def purity_score(y_true, y_pred):
    contingency_matrix = metrics.cluster.contingency_matrix(y_true, y_pred)
    return np.sum(np.amax(contingency_matrix, axis=0)) / np.sum(contingency_matrix)
```

```python
purity_score(df["Y2"],df["Y_true"])
```
